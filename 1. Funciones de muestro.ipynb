#### 2) Muestro sistemático

```python
def systematic_sampling(econdata,step):
    indexes = np.arange(0,len(econdata), step=step)
    systematic_sampling = econdata.iloc[indexes]
    return systematic_sampling

systematic_sample = systematic_sampling(econdata,3)
systematic_sample
```

```python
econdata['estratificado'] = econdata ['delegacion'] + ',' + econdata['tipo']
(econdata['estratificado'].value_counts()/len(econdata)).sort_values(ascending=False)
```

Nuestros datos dicen que la proporción es la siguiente:
1) Hoteles en Cauhtémoc 0.5
2) Museos en Cauhtémoc 0.2
3) Hoteles en Venustiana Carranza 0.1
4) Mercados en Cuautémoc 0.1

```python
# Definir tamaño de la muestra deseada:

tasa_muestra = 0.20
tamaño_muestra = int(len(econdata) * tasa_muestra)
```

```python
def data_estratificada(econdata, nombres_columnas_estrat, valores_estrat, prop_estrat, tamaño_muestra, random_state=None):
    df_estrat = pd.DataFrame(columns=econdata.columns)

    for i in range(len(valores_estrat)):
        if i == len(valores_estrat) - 1:
            ratio_len = tamaño_muestra - len(df_estrat)
        else:
            ratio_len = int(tamaño_muestra * prop_estrat[i])

        df_filtrato = econdata[econdata[nombres_columnas_estrat] == valores_estrat[i]]
        df_temp = df_filtrato.sample(n=ratio_len, random_state=random_state)

        df_estrat = pd.concat([df_estrat, df_temp], ignore_index=True)

    return df_estrat
```

```python
valores_estrat = ['Cuautémoc,Hotel', 'Cuautémoc,Museo', 'Venustiano Carranza,Hotel', 'Cuauhtémoc,Mercado', 'Venustiano Carranza,Mercado']
prop_estrat = [0.5, 0.2, 0.1, 0.1, 0.1]
df_estrat = data_estratificada(econdata, 'estratificado', valores_estrat, prop_estrat, tamaño_muestra, random_state=42)
```
